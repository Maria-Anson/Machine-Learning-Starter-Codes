# Sample plot with generated dataset

```python
z = np.linspace(0,15,1000)
y = np.cos(z)
x = np.sin(z)

fig =plt.figure(figsize=(10,10))
ax = plt.axes(projection = '3d')

ax.plot3D(x,y,z)
```

```python
z = np.linspace(0,15,1000)
y = np.cos(z)
x = np.sin(z)

fig =plt.figure(figsize=(10,10))
ax = plt.axes(projection = '3d')

ax.plot3D(x,y,z,'red')

zpoint = 15 * np.random.random(100)
ypoint = np.cos(zpoint) + 0.1*np.random.random(100)
xpoint = np.sin(zpoint) + 0.1*np.random.random(100)
ax.scatter3D(xpoint,ypoint,zpoint)
```

# Contour plot - Continuous value for all

```python
x = np.array([1,2,3,4])
y = np.array([5,6,7])

x,y = np.meshgrid(x,y)

print(x,'\n',y)
```

```python
y = np.linspace(0,5,300)
x = np.linspace(0,5,300)

x,y = np.meshgrid(x,y)

z = x**2 + y**2

fig =plt.figure(figsize=(10,10))
ax = plt.axes(projection = '3d')

ax.contour3D(x,y,z,50)
```

```python
y = np.linspace(-5,5,300)
x = np.linspace(-5,5,300)

x,y = np.meshgrid(x,y)

z = np.sin(np.sqrt(x**2 + y**2))

fig =plt.figure(figsize=(10,10))
ax = plt.axes(projection = '3d')

ax.contour3D(x,y,z,50,cmap='binary')
```

```python
#Surface Triangulation
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))


theta = 2 * np.pi * np.random.random(1000)
r = 6 * np.random.random(1000)
x = np.ravel(r * np.sin(theta))
y = np.ravel(r * np.cos(theta))
z = f(x, y)
plt.figure(figsize=(10,10))

ax = plt.axes(projection='3d')
ax.scatter(x, y, z, c=z, cmap='viridis', linewidth=2);
```
